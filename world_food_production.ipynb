### Top ten most food production country in each fodd item

```python
# Step 1: Aggregate the data by Entity (sum production over all years)
aggregated_df = df.groupby('Entity').sum().reset_index()

# Step 2: Plot the top 10 producers for each type of food
for column in plotting_column:
    # Sort the aggregated dataframe by the current column in descending order and select the top 10 producers
    top_10_df = aggregated_df.sort_values(by=column, ascending=False).head(10)
    
    # Generate the bar plot for the top 10 producers
    fig = px.bar(top_10_df, x='Entity', y=column, title=f'Top 10 Producers - {column}',
                    text=top_10_df[column].apply(lambda x:f'{x:.5f}'))
    
    fig.update_traces(textposition='outside')
    # Show the plot
    fig.show()
```

## Outliers detetion

```python
num_feature =  df.select_dtypes(include='number')
```

### Percentage of outliers by interqurtile range

```python
print("Percentage of outliers")
print()
outliers_percentage = {}
all_outliers = set() # To store unique outlier indices
for col in num_feature:
    # Access the specific column as a Series
    column_data = df[col]
    
    # Calculate Q1 and Q3
    Q1 = column_data.quantile(0.25)
    Q3 = column_data.quantile(0.75)

    # Calculate the IQR
    IQR = Q3 - Q1
    
    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers specific to the column
    outliers = column_data[(column_data < lower_bound) | (column_data > upper_bound)]
    
    #  Add outlier indices to the set of all outliers
    all_outliers.update(outliers.index)

    # Calculate the percentage of outliers
    outlier_percentage = round((len(outliers) / len(df)) * 100,2)

    # Store the result in the dictionary
    outliers_percentage[col] = outlier_percentage
    print(col,' = ',outlier_percentage,'%')

# calculate the overall percentage of unique outliers
total_outliers = len(all_outliers)
overall_percentage = round((total_outliers / len(df))*100,2)
print()
print()
print("Percentage of total outliers ","=",overall_percentage,'%')
```
